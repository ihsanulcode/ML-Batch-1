## Load Dataset

```python
import pandas as pd

df = pd.read_csv("iris.csv")
```

# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Count Plot

```python
sns.countplot(data=df,y='species')
```

```python
sns.barplot(data=df,x='species',y='sepal_length')
plt.show()
```

```python
sns.swarmplot(data=df,x='species',y='petal_length')
```

```python
# Set the plot and fig size
plt.figure(figsize=(20,10))

# count plot
sns.countplot(data=df, x='petal_length', hue='species')

# Labeling
plt.ylabel('Count per Species')
plt.title('Iris Flower Species Analysis')

# Show plot
plt.show()
```

```python
# Box Plot
sns.boxplot(x='species',y='sepal_length',data=df)
plt.show()
```

```python
# Pair Plot

sns.pairplot(df,hue='species')
plt.show()
```

```python
# Hist plot

sns.histplot(data=df,x='sepal_length')
# plt.grid()
plt.show()
```

```python
# Hist plot

sns.histplot(data=df,x='sepal_length')
plt.grid()
plt.show()
```

```python
# Histograms together
df.hist(edgecolor='black',linewidth=1.2)
plt.tight_layout()
plt.show()
```

```python
# Distribution plot

sns.distplot(df.petal_length,color='y')
```
